1. Las columnas `land_area`, `armed_forces_size`, `co2-emissions`, `gdp`, `population`, `urban_population`, `latitude`, `longitude`, `density`, `cpi` son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
#Vemos que gdp, population y urban_population tienen varias ',' como separadores de miles (no como separación decimal) así que hacemos una función que tenga en cuenta cuando hay 1 ',' o más de una.
```

```python
#Creamos la función
def object_to_float(cadena):
    try:
        #si hay más de 1 ',' es un separador de miles, por lo que no ponemos '.'
        if cadena.count(',') > 1:
            return round(float(cadena.replace(',', '')),2)
        #si tiene una sola ',' es un separador decimal, por lo que sí ponemos '.'
        elif cadena.count(',') == 1:
            return round(float(cadena.replace(',', '.')),2)
        #si no tiene ninguna ',' lo pasamos directamente a float (no tenía decimales)
        else:
            return round(float(cadena),2)
    except:
        return np.nan
```

```python
#Aplicamos la función
col_object_to_float=['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitude', 'longitude', 'density', 'cpi']
for i in col_object_to_float:
    df_world_continent[i]=df_world_continent[i].apply(object_to_float)

#Comprobamos resultado
df_world_continent.head(2)
```
