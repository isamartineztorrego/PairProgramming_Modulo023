```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
#Ponemos index_col=0 para que no traiga la columna 'Unnamed'
df_world1 = pd.read_csv("world-data-2023_part1.csv", index_col=0)
df_world2 = pd.read_csv("world-data-2023_part2.csv", index_col=0)
```

```python
df_world1.head(2)
```

```python
df_world2.head(2)
```

```python
#Unimos con un merge por la columna común 'country', y añadiendo las columnas del df_world2 a la derecha del df_world1, así tenemos un único df con todas las columnas
df_world=pd.merge(left=df_world1, right=df_world2, left_on='Country', right_on='country', how='inner')
df_world.head(2)
```

#### Ejercicios de Limpieza
1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
#Eliminamos columna sobreeescribiendo el resultado
df_world.drop(['country'], axis=1, inplace=True)
df_world.head(1)
```

❗Antes de nada, queremos eliminar los posibles espacios en blanco que pueda haber al final del nombre de cada columna. Para ello, hacemos un strip()

```python
df_world.columns=df_world.columns.str.rstrip()
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

- No tengan espacios.

```python
#Iteramos las columnas del df_merged y reemplazamos los ' ' por '_'
#for i in df_merged.columns:
#   i.replace(' ', '')

df_world.columns=[i.replace(' ','_') for i in df_world.columns]
df_world.head(1)
```

- Estén en minúscula.

```python
df_world.columns=[i.lower() for i in df_world.columns]
df_world.head(1)
```

- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

```python
#Hacemos un un patrón regex que:
#Identifique los paréntesis exactos --> \( \)
#Identifique cualquier carácter salvo salto de línea --> .
#Que ese carácter anterior esté 0 o más veces --> *
#Limita el * a no codicioso, es decir, que coincida con la menor cantidad de caracteres posible (esto permite coincidencias más precisas y evita abarcar más de lo necesario)
df_world.columns=df_world.columns.str.replace(r'\(.*?\)', '', regex=True).str.strip()
df_world.head(1)
```

- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

```python
df_world.columns=[i.replace('\n', '') for i in df_world.columns]
df_world.head(1)
```

- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df_world.columns=[i.replace(':', '') for i in df_world.columns]
df_world.head(1)
```

❗Ahora vemos que las columnas en las que quitamos los () terminan en _ (porque antes era un espacio en blanco y lo sustituimos por _). Vamos a quitar esos _ finales.

```python
df_world.columns=df_world.columns.str.rstrip('_')
df_world.head(1)
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de `coordinates`.

```python
#Creamos las nuevas columnas y separamos la columna coordinates por ','. Con expand=True le dedimos que el resultado del split lo expanda en las nuevas columnas.
df_world[['longitude', 'latitude']]=df_world['coordinates'].str.split(',', expand=True)
df_world.head(1)
```

```python
#Limpiamos el resultado para que únicamente queden los valores, sin () ni '
df_world['longitude']=df_world['longitude'].str.replace('(', '').str.replace("'", "")
df_world['latitude']=df_world['latitude'].str.replace(')', '').str.replace("'", "")
df_world.head(1)
```

```python
#Eliminamos columna coordinates
df_world.drop('coordinates', axis=1, inplace=True)
df_world.head(1)
```

4. Las columnas `unemployment_rate`, `total_tax_rate`, `tax_revenue`, `population_labor_force_participation`, `out_of_pocket_health_expenditure`, `gross_tertiary_education_enrollment`, `gross_primary_education_enrollment`, `forested_area`, `cpi_change`, `agricultural_land` tienen "%". Elimina los "%" de los valores de las columnas.

```python
columns_percentage=['unemployment_rate', 'total_tax_rate', 'tax_revenue', 'population_labor_force_participation', 'out_of_pocket_health_expenditure', 'gross_tertiary_education_enrollment', 'gross_primary_education_enrollment', 'forested_area', 'cpi_change', 'agricultural_land']
for i in columns_percentage:
    df_world[i]=df_world[i].str.replace('%', '')

df_world.head(1)
```

5. Haz lo mismo para las columnas de `gasoline_price`, `gdp`, `minimum_wage`, pero eliminando "$".

```python
columns_money=['gasoline_price', 'gdp', 'minimum_wage']
for i in columns_money:
    df_world[i]=df_world[i].str.replace('$', '')

df_world.head(1)
```

6. Encuentra los países cuya moneda es el euro (EUR) y tienen una tasa de natalidad superior al promedio.

```python
#Queremos ver sólo las columnas country, currency-code, birth_rate
#1. Calculamos la media de la natalidad
mean_natality=df_world['birth_rate'].mean()

#2. Creamos filtros
filter_currency=df_world['currency-code'] =='EUR'
filter_natality=df_world['birth_rate'] > mean_natality

#3. Seleccionamos las columnas que necesitamos en base a los filtros
filter_currency_natality=df_world.loc[(filter_currency & filter_natality), ['country', 'birth_rate', 'currency-code']]
filter_currency_natality
```
